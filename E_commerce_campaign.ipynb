# Hypothesis Testing

## Is income of customers dependent on their education ?

```python
df['Income'] = df['Income'].str.replace('$nan', '0')
df['Income'] = df['Income'].str.replace('$', '')
df['Income'] = df['Income'].str.replace(',', '')
```

```python
df['Income'] = df['Income'].astype(str).astype(float)

df.head()
```

```python
df["Income"].max()
```

```python
# Creating Income_bin feature

bins=[0,40000,80000,120000,160000,200000]
group=["Low",'Middle Class','Upper_Middle Class','Rich','Super_rich']

df['IncomeBins']= pd.cut(df['Income'],bins,labels=group)
df.head(2)
```

```python
pd.crosstab(df["IncomeBins"], df["Education"])
```

```python
# Test to check dependancy between Income_bin and Education
from scipy.stats import chi2_contingency

val = pd.crosstab(index=df["IncomeBins"], columns=df["Education"]).values
print(val)
print()

chi2_contingency(val) # chi_stat, p_value, df, expected_values

chi_stat, p_value, df, exp_freq = chi2_contingency(val)

print(p_value)
print()

if p_value > 0.05:
  print("Failed to reject null hypothesis")
  print("Income and Education are independent")
else:
  print("Reject null hypothesis")
  print("Income and Education are dependent")
```

## Do higher income people spend more (take in account spending in all categories together)

```python
from google.colab import files
files = files.upload()
```

```python
df = pd.read_csv("campaign - campaign.csv")
df.head()
```

```python
df['Income'] = df['Income'].str.replace('$nan', '0')
df['Income'] = df['Income'].str.replace('$', '')
df['Income'] = df['Income'].str.replace(',', '')
df['Income'] = df['Income'].astype(str).astype(float)

df.head()
```

```python
df["Total_spend"] = df["MntWines"] + df["MntFruits"] + df["MntMeatProducts"] + df["MntFishProducts"] + df["MntSweetProducts"] + df["MntGoldProds"]
```

```python
# Creating Income_bin feature

bins=[0,30000,60000,90000,120000,150000,180000]
group=["Lower","Average",'Middle Class','Upper_Middle Class','Rich','Super_rich']

df['IncomeBins']= pd.cut(df['Income'],bins,labels=group)
```

```python
a = df.loc[(df["IncomeBins"]=="Rich") | (df["IncomeBins"]== "Super_Rich") | (df["IncomeBins"]== "Upper_Middle Class")]["Total_spend"]

b = df.loc[(df["IncomeBins"]=="Low") | (df["IncomeBins"]== "Middle Class") | (df["IncomeBins"]== "Average")]["Total_spend"]
```

Hypothesis

* H0 : No difference between spending of rich and less rich
* Ha : There is significant difference between spending of rich and less rich.
* Significance level 95 %

```python
from scipy.stats import ttest_ind
t_stat, p_value = ttest_ind(a, b, alternative = "greater")
print(p_value)
if p_value < 0.05:
    print("Reject H0")
    print("There is significant difference between spending of rich and less rich")
else:
    print("Fail to reject H0")
    print("No difference between spending of rich and less rich")
```

## Do couples spend more or less money on wine than people living alone (set 'Married','Together':'In couple' and 'Divorced','Single','Absurd','Widow','YOLO':'Alone')

```python
a = df.loc[(df["Marital_Status"]=="Couple") | (df["Marital_Status"]== "Together") ]["MntWines"]

b = df.loc[(df["Marital_Status"]=="Divorced") | (df["Marital_Status"]== "Single")| (df["Marital_Status"]== "Absurd") |(df["Marital_Status"]== "Yolo") | (df["Marital_Status"]== "Alone") | (df["Marital_Status"]== "Widow") ]["MntWines"]
```

Hypothesis

* H0 : No difference between spending of couples and alone on wine.
* Ha : There is significant difference between spending of couple and alone for wine.
Significance level 95 %

```python
from scipy.stats import ttest_ind
t_stat, p_value = ttest_ind(a, b, alternative = "two-sided")
print(p_value)
if p_value < 0.05:
    print("Reject H0")
    print("There is significant difference between spending of couple and alone for wine. Significance level 95 %")
else:
    print("Fail to reject H0")
    print("No difference between spending of couples and alone on wine.")
```

Hypothesis

* Significance level 95 %

```python
from scipy.stats import ttest_ind
t_stat, p_value = ttest_ind(Upper_income,Lower_income, alternative = "less")
print(p_value)
if p_value < 0.05:
    print("Reject H0")
    print("There is significant difference accepting campaign offers for people of Lower income and upper income bracket.")
else:
    print("Fail to reject H0")
    print("No difference between accepting campaign offers for people of Lower income and upper income bracket.")
```
